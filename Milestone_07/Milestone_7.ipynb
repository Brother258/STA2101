## A. Introduction

This milestone studies the linear relationship between two numerical variables. I calculated regression parameters manually using formulas. No machine learning library is used.

## B. Knowledge Points: Least Squares Method

I used the least squares formulas for slope, intercept, correlation, and R². All values are computed step by step.

## C. Task 1: Data Selection and Initial Visualization

Independent variable (X): temperature (temp_c)

Dependent variable (Y): humidity

These two variables are expected to have a relationship.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

DATA_PATH = r"2025-08-10.csv"
df = pd.read_csv(DATA_PATH)

X = df['temp_c'].dropna()
Y = df.loc[X.index, 'humidity']

X_bar = X.mean()
Y_bar = Y.mean()
sx = X.std()
sy = Y.std()

print('Mean of X (temp):', round(X_bar,3))
print('Mean of Y (humidity):', round(Y_bar,3))
print('Std of X:', round(sx,3))
print('Std of Y:', round(sy,3))

plt.figure(figsize=(6,4))
plt.scatter(X, Y)
plt.xlabel('Temperature (temp_c)')
plt.ylabel('Humidity')
plt.title('Scatter Plot: Humidity vs Temperature')
plt.show()
```

## D. Task 2: Manual Calculation of Regression Parameters

I computed slope and intercept using formulas only.

```python
num = ((X - X_bar) * (Y - Y_bar)).sum()
den = ((X - X_bar)**2).sum()

beta1 = num / den
beta0 = Y_bar - beta1 * X_bar

print('Numerator Σ(X−X̄)(Y−Ȳ) =', round(num,3))
print('Denominator Σ(X−X̄)² =', round(den,3))
print('Slope β1 =', round(beta1,4))
print('Intercept β0 =', round(beta0,4))

print('Regression Equation:')
print('Ŷ =', round(beta0,3), '+', round(beta1,3), '* X')
```

## E. Task 3: Visualization of the Fit and Interpretation

I plotted the regression line and explained slope meaning.

```python
Y_hat = beta0 + beta1 * X

plt.figure(figsize=(6,4))
plt.scatter(X, Y)
plt.plot(X, Y_hat)
plt.xlabel('Temperature (temp_c)')
plt.ylabel('Humidity')
plt.title('Regression Line on Scatter Plot')
plt.show()

print('Interpretation:')
print('When temperature increases by 1 unit, humidity changes by about', round(beta1,3))
```

## F. Task 4: Strength of Relationship

I computed correlation and R² manually.

```python
r_num = ((X - X_bar) * (Y - Y_bar)).sum()
r_den = np.sqrt(((X - X_bar)**2).sum() * ((Y - Y_bar)**2).sum())
r = r_num / r_den
R2 = r**2

print('Pearson r =', round(r,4))
print('R² =', round(R2,4))

if abs(r) < 0.3:
    strength = 'weak'
elif abs(r) < 0.6:
    strength = 'moderate'
else:
    strength = 'strong'

direction = 'positive' if r > 0 else 'negative'
print('Relationship:', strength, 'and', direction)
print('About', round(R2*100,2), '% of humidity variation is explained by temperature')
```

## G. Task 5: Reflection

The reflection is printed below.

```python
print('Reflection:')
print('The regression line follows the data points reasonably.')
print('R² supports the visual fit.')
print('This model can help predict humidity from temperature.')
```
